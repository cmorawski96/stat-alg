<h4>Dla zmiennych losowych przedstawionych w tabeli obliczyć podstawowe statystyki
    
|Wartość|1|2|3|4|5|6|
|---|---|---|---|---|---|---|
|Prawdopodobieństwo|1/6|1/6|1/6|1/6|1/6|1/6
    </h4>

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import statistics as stat
import scipy.stats as scs
import seaborn as sns
```

```python
elements = (1, 2, 3, 4, 5, 6)
probabilities = (1/6,1/6,1/6,1/6,1/6,1/6)
data = scs.rv_discrete(name="firstVar", values=(elements,probabilities))
print("Mean: ", data.mean())
print("Variance: ", data.var())
print("Standard deviation: ",data.std())
print("Median: ",data.median())
for i in range(7):
    print("Dystrybuanta (cumulative distribution function): ", i," - ", data.cdf(i))
print("Expected value: ", data.expect())
```

<h4><hr>Wygeneruj próby dla n=100 dla następujących rozkładów: Bernoulliego, Dwumianowego, Poissona</h4>

```python
p = 0.3
n = 5
bernoulli = scs.bernoulli.rvs(p, size=100)
binomial = scs.binom.rvs(n, p, size=100)
poisson = scs.poisson.rvs(p, size=100)
```

<hr><h4>Policz dla nich statystyki podstawowe (średnią, wariancję, kurtozę i skośność).</h4>

```python
mean, var, skew, kurt = scs.bernoulli.stats(p, moments='mvsk')
bernMean = mean
bernVar = var
bernKurt = kurt
bernSkew = skew

mean, var, skew, kurt = scs.binom.stats(n, p, moments='mvsk')
binMean = mean
binVar = var
binKurt = kurt
binSkew = skew

mean, var, skew, kurt = scs.poisson.stats(p, moments='mvsk')
poiMean = mean
poiVar = var
poiKurt = kurt
poiSkew = skew
```

```python
print('\nBernoulli')
print('Mean: ', bernMean)
print('Variance: ',bernVar)
print('Skewness: ',bernSkew)
print('Kurtosis: ',bernKurt)

print('\nBinomial')
print('Mean: ', binMean)
print('Variance: ',binVar)
print('Skewness: ',binSkew)
print('Kurtosis: ',binKurt)

print('\nPoisson')
print('Mean: ', poiMean)
print('Variance: ',poiVar)
print('Skewness: ',poiSkew)
print('Kurtosis: ',poiKurt)

```

<hr><h4>Wygeneruj 100 danych dla rozkładu normalnego o średniej równej 0 i odchyleniu standardowym równym 2.
Wyznacz wszystkie statystyki podstawowe – czy są one równe z wartościami teoretycznymi? Sprawdź, czy
zwiększenie liczby danych zwiększy dokładność wyliczeń statystyk opisowych.</h4>

```python
#loc = mean, scale = std
data = scs.norm.rvs(size=10000, loc=0, scale=2)
mean, var, skew, kurt = scs.norm.stats(loc=0, scale=2, moments='mvsk')
print("Mean: should be 0, is ", mean)
print("Variance: should be 4, is ", var)
print("Standard deviation: should be 2, is ", data.std())
print("Median: should be 0, is",scs.norm.median(loc=0, scale=2))
print("Expected value: should be 0, is ", scs.norm.expect(loc=0, scale=2))
print('Kurtosis: should be 0, is ', kurt)
print('Skewness: should be 0, is ', skew)
```

Zwiekszenie liczby danych zwieksza dokladnosc statystyk, lecz w nieznacznym stopniu powyzej 100

<hr><h4>Narysować na jednym wykresie histogram dla rozkładu normalnego o parametrach: średnia = 1, odchylenie
=2, wykres dla rozkładu standardowego, oraz wykres gęstości dla średniej równej -1 oraz odchylenia
równego 0.5.</h4>

```python
data1 = scs.norm.rvs(size=1000, loc=1, scale=2)
data2 = scs.norm.rvs(size=1000, loc=0, scale=1)
data3 = scs.norm.rvs(size=1000, loc=-1, scale=0.5)

ax = sns.distplot(data1,color='olive')
ax = sns.distplot(data2,color='fuchsia')
ax = sns.kdeplot(data3,color='skyblue')
ax.legend()
```
